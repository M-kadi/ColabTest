```python
nums = list(range(0, 1000001))
len(nums)
```

```python
nums_rdd = sc.parallelize(nums)
nums_rdd
```

```python
nums_rdd.collect()
```

```python
nums_rdd.take(5)
```

```python
3 ** 2
```

```python
squared_nums_rdd = nums_rdd.map(lambda x: x ** 2)
squared_nums_rdd.take(5)
```

```python
len(str(546))
```

```python
pairs = squared_nums_rdd.map(lambda x: (x, len(str(x))))
pairs.take(25)
```

```python
even_digit_pairs = pairs.filter(lambda x: (x[1] % 2) == 0)
even_digit_pairs.take(25)
```

```python
flipped_pairs = even_digit_pairs.map(lambda x: (x[1], x[0]))
flipped_pairs.take(25)
```

```python
grouped = flipped_pairs.groupByKey()
grouped.take(25)
```

```python
listStr = ['Ahmad', 'Mohammad', 'Car', 'Ahmad', 'Car', 'Samer', 'Mohammad']
```

```python
listStr_rdd = sc.parallelize(listStr)
listStr_rdd
```

```python
listStr_rdd.take(5)
```

```python
map = listStr_rdd.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1))
mapCount = map.reduceByKey(lambda x, y: x+y)
mapCount.collect()
```

```python
from operator import add
rdd = sc.parallelize([("a", 3), ("b", 1), ("a", 1)])
sorted(rdd.reduceByKey(add).collect())
```
